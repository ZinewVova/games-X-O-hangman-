# Крестики-нолики

```python
def field(cells):
  print('_'*13)
  print(f'| {cells[0]} | {cells[1]} | {cells[2]} |')
  print('-'*13)
  print(f'| {cells[3]} | {cells[4]} | {cells[5]} |')
  print('-'*13)
  print(f'| {cells[6]} | {cells[7]} | {cells[8]} |')
  print('-'*13)
```

```python
def turnf(cells, turn):
    print('\n', '_'*12)
    print(f'\n   ходит {turn}\n')
    field(cells)
    choice = int(input(f'Поставить {turn} в ячейку №'))
    cells[choice - 1] = f'{turn}'

    return cells
```

```python
def rules(cells):
    win = [[0, 1, 2], [3, 4, 5], [6, 7, 8],
           [0, 3, 6], [1, 4, 7], [2, 5, 8],
          [0, 4, 8], [2, 4, 6]]
    
    for combination in win:
        symbols = [cells[i] for i in combination]
        if symbols == ['X', 'X', 'X']:
            print('\nПобедил игрок играющий за Х')
            field(cells)
            return True
        elif symbols == ['O', 'O', 'O']:
            print('\nПобедил игрок играющий за O')
            field(cells)
            return True
    
    return False
```

```python
def change_turn(turn):
    if turn == 'X':
        return'O'
    else:
        return'X'
```

```python
def game():
  turn = 'X'
  cells = [i for i in range(1,10)]
  move = 0
  while not rules(cells): 
      if move == 9:
          print('Ничья')
          break
      cells = turnf(cells, turn) 
      turn = change_turn(turn)
      move+=1
```

```python
game()
```

## создание игры

```python
import random
```

```python
# случайный выбор слова
def get_word():
    word = random.choice(words)
    return word.upper()

# игра
def play(word):
    word_completion = '_' * len(word)
    guessed = False
    guessed_letters = []
    guessed_words = []
    tries = 8
    print("let's play hangman!")
    print(display_hangman(tries))
    print(' '.join(word_completion))
    print("\n")
    while not guessed and tries > 0:
        guess = input("please guess a letter or word: ").upper()
        if len(guess) == 1 and guess.isalpha():
            if guess in guessed_letters:
                print("you already guessed the letter", guess)
            elif guess not in word:
                print(guess, "is not in the word.")
                tries -= 1
                guessed_letters.append(guess)
            else:
                print("good job,", guess, "is in the word!")
                guessed_letters.append(guess)
                word_as_list = list(word_completion)
                indices = [i for i, letter in enumerate(word) if letter == guess]
                for index in indices:
                    word_as_list[index] = guess
                word_completion = word_as_list
                if "_" not in word_completion:
                    guessed = True
        elif len(guess) == len(word) and guess.isalpha():
            if guess in guessed_words:
                print("you already guessed the word", guess)
            elif guess != word:
                print(guess, "is not the word.")
                tries -= 1
                guessed_words.append(guess)
            else:
                guessed = True
                word_completion = word
        else:
            print("not a valid guess.")
        print(display_hangman(tries))
        print(' '.join(word_completion))
        print("\n")
    if guessed:
        print("congratulations, you guessed the word!")
    else:
        print(f"sorry, you ran out of tries. The word was {word}.")

# отрисовка
def display_hangman(tries):
    stages = ["""
                   --------
                   |      |
                   |      0
                   |     /|\\
                   |      |
                   |     / \\
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / 
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |      
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """,
              """
                   --------
                   |
                   |  
                   |
                   | 
                   |
                   -
              """
    ]
    return stages[tries]
```

```python
def hangman():
    word = get_word()
    play(word)
    while input("wanna play again? (yes/no) ").lower() == "yes":
        word = get_word()
        play(word)
```
